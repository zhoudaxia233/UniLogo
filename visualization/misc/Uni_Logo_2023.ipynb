```python
import time
from pathlib import Path
from urllib.parse import urljoin

import requests
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
current_folder = Path.cwd()
```

```python
base_url = 'https://www.shanghairanking.com'
url = urljoin(base_url, 'rankings/arwu/2022')
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
img_page = soup.select('.univ-logo')
img_page
```

```python
folder_path = current_folder / Path("images")

if not folder_path.exists():
    folder_path.mkdir()
    print(f"Folder 'images' created.")
else:
    print(f"Folder 'images' already exists.")
```

```python
rank = []
for i in tqdm(range(30)):
  uni_name, short_path = img_page[i]['alt'], img_page[i]['src']
  rank.append(uni_name)

  rsp = requests.get(urljoin(base_url, short_path))

  with open(current_folder / "images" / f"{uni_name}{Path(short_path).suffix}", 'wb') as f:
    f.write(rsp.content)
```

```python
rank_df = pd.DataFrame({'Rank': range(1, 1 + len(rank)), 'University': rank})
rank_df.to_excel(current_folder / 'uni ranking.xlsx', index=False)
rank_df
```
